```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

### Análisis de Correlación entre Columnas Numéricas

Vamos a calcular la matriz de correlación para las columnas numéricas en el DataFrame y visualizarla con un mapa de calor para identificar relaciones entre las variables.

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Proyecto 3 Quimica Organica.xlsx")

# Limpiar y convertir la columna 'PE_Sustrato °C' a numérica
# Esto convertirá valores no numéricos (como '219-223 ') a NaN
df['PE_Sustrato °C'] = pd.to_numeric(df['PE_Sustrato °C'], errors='coerce')

# Eliminar filas con valores NaN en las columnas que usaremos para el boxplot
df_cleaned = df.dropna(subset=["PE_Sustrato °C", "Grupo Funcional"])

# Crear gráfico boxplot
df_cleaned.boxplot(column="PE_Sustrato °C", by="Grupo Funcional", figsize=(12, 8))

plt.title("Grupo Funcional vs Punto de Ebullición")
plt.xlabel("Grupo Funcional")
plt.ylabel("PE Sustrato °C")
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Proyecto 3 Quimica Organica.xlsx")

# Tabla de frecuencia usando el nombre de columna correcto 'Catalizador'
datos = df.groupby(["Catalizador", "Tipo_Reaccion"]).size().unstack(fill_value=0)

# Graficar
datos.plot(kind="bar", figsize=(12, 7))

plt.title("Frecuencia de Tipos de Reacción por Catalizador")
plt.xlabel("Catalizador")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45, ha='right') # Rotar las etiquetas para mejor legibilidad
plt.tight_layout() # Ajustar el diseño para evitar superposiciones

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Proyecto 3 Quimica Organica.xlsx")

# Limpiar y convertir las columnas a numéricas, manejando posibles errores
df['PM_Sustrato'] = pd.to_numeric(df['PM_Sustrato'], errors='coerce')
df['PM_Producto'] = pd.to_numeric(df['PM_Producto'], errors='coerce')

# Eliminar filas con valores NaN en las columnas que usaremos para el gráfico
df_cleaned = df.dropna(subset=["PM_Sustrato", "PM_Producto"])

# Seleccionar variables de la DataFrame limpia
x = df_cleaned["PM_Sustrato"]
y = df_cleaned["PM_Producto"]

# Crear gráfico de dispersión
plt.figure(figsize=(10, 7))
plt.scatter(x, y)

# Títulos y etiquetas
plt.title("PM_Sustrato vs PM_Producto")
plt.xlabel("Peso Molecular del Sustrato (PM_Sustrato)")
plt.ylabel("Peso Molecular del Producto (PM_Producto)")
plt.grid(True)

# Mostrar gráfico
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Proyecto 3 Quimica Organica.xlsx")

# Limpiar y convertir la columna 'PE_Producto' a numérica
df['PE_Producto'] = pd.to_numeric(df['PE_Producto'], errors='coerce')

# Eliminar filas con valores NaN en las columnas que usaremos
df_cleaned = df.dropna(subset=["Nombre_Solvente", "PE_Producto"])

# Promedio de PE por solvente utilizando el nombre de columna correcto
datos = df_cleaned.groupby("Nombre_Solvente")["PE_Producto"].mean()

datos.plot(kind="bar")

plt.title("Solvente vs PE del Producto")
plt.xlabel("Solvente")
plt.ylabel("PE Producto")
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("Proyecto 3 Quimica Organica.xlsx")

# Limpiar y convertir las columnas a numéricas, manejando posibles errores
df['PE_Sustrato °C'] = pd.to_numeric(df['PE_Sustrato °C'], errors='coerce')
df['PE_Producto'] = pd.to_numeric(df['PE_Producto'], errors='coerce')

# Eliminar filas con valores NaN en las columnas que usaremos para el gráfico
df_cleaned = df.dropna(subset=["PE_Sustrato °C", "PE_Producto"])

x = df_cleaned["PE_Sustrato °C"]
y = df_cleaned["PE_Producto"]

# Gráfico de dispersión
plt.scatter(x, y)

# Línea de tendencia (solo si hay suficientes datos para el ajuste)
if len(x) > 1:
    m, b = np.polyfit(x, y, 1)
    plt.plot(x, m*x + b, color='red', linestyle='--', label='Línea de Tendencia')
    plt.legend()
else:
    print("No hay suficientes datos para calcular y mostrar una línea de tendencia.")

plt.title("PE Sustrato vs PE Producto")
plt.xlabel("PE Sustrato °C")
plt.ylabel("PE Producto")

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Proyecto 3 Quimica Organica.xlsx")

# Limpiar y convertir la columna 'PE_Producto' a numérica
# Esto convertirá valores no numéricos a NaN
df['PE_Producto'] = pd.to_numeric(df['PE_Producto'], errors='coerce')

# Eliminar filas con valores NaN en las columnas que usaremos para el boxplot
df_cleaned = df.dropna(subset=["PE_Producto", "Grupo Funcional"])

# Crear gráfico boxplot
plt.figure(figsize=(12, 8))
df_cleaned.boxplot(column="PE_Producto", by="Grupo Funcional")

plt.title("Grupo Funcional vs Punto de Ebullición del Producto")
plt.xlabel("Grupo Funcional")
plt.ylabel("PE Producto")
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.suptitle('') # Remove default suptitle to avoid overlap with plt.title
plt.tight_layout() # Adjust layout to prevent labels from overlapping

plt.show()
```
