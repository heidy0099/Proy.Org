```python
# ============================================
# ANALISIS 2: Boxplot por Grupo Funcional - VERSIÓN FLEXIBLE
# ============================================

import plotly.express as px

# --------------------------------------------
# BLOQUE 1: ENCONTRAR COLUMNAS AUTOMÁTICAMENTE
# --------------------------------------------
col_grupo = None
col_PE = None

for col in df.columns:
    if 'Grupo' in col or 'Funcional' in col or 'GF' in col:
        col_grupo = col
    if 'PE' in col or 'Ebullicion' in col or 'PES' in col:
        col_PE = col

print(f"🔍 Columna Grupo Funcional encontrada: {col_grupo}")
print(f"🔍 Columna PE encontrada: {col_PE}")

# --------------------------------------------
# BLOQUE 2: PARÁMETROS MODIFICABLES
# --------------------------------------------
PE_MIN = 20        # Temperatura mínima
PE_MAX = 400       # Temperatura máxima
MIN_COMPUESTOS = 3 # Mínimo compuestos por grupo

if col_grupo and col_PE:
    # --------------------------------------------
    # BLOQUE 3: LIMPIAR Y FILTRAR DATOS
    # --------------------------------------------
    plot_df = df[[col_grupo, col_PE]].dropna()
    plot_df = plot_df[(plot_df[col_PE] >= PE_MIN) & (plot_df[col_PE] <= PE_MAX)]

    # Filtrar grupos con suficientes datos
    grupos_validos = plot_df[col_grupo].value_counts()
    grupos_validos = grupos_validos[grupos_validos >= MIN_COMPUESTOS].index
    plot_df = plot_df[plot_df[col_grupo].isin(grupos_validos)]

    print(f"✅ Grupos analizados: {list(plot_df[col_grupo].unique())}")
    print(f"✅ Total compuestos: {len(plot_df)}")

    # --------------------------------------------
    # BLOQUE 4: CREAR BOXPLOT
    # --------------------------------------------
    fig = px.box(
        plot_df,
        x=col_grupo,
        y=col_PE,
        color=col_grupo,
        title=f'Punto de Ebullición por {col_grupo}',
        labels={col_grupo: 'Familia Química', col_PE: 'Temperatura (°C)'},
        points='outliers'  # 'all', 'outliers', False
    )

    # --------------------------------------------
    # BLOQUE 5: AJUSTAR DISEÑO
    # --------------------------------------------
    fig.update_layout(
        width=900,
        height=550,
        xaxis_tickangle=-45,
        showlegend=False,
        plot_bgcolor='white'
    )

    fig.show()

    # --------------------------------------------
    # BLOQUE 6: MOSTRAR ESTADÍSTICAS
    # --------------------------------------------
    print("\n" + "="*50)
    print("ESTADÍSTICAS POR GRUPO FUNCIONAL")
    print("="*50)

    for grupo in plot_df[col_grupo].unique():
        datos = plot_df[plot_df[col_grupo] == grupo][col_PE]
        print(f"\n📌 {grupo}:")
        print(f"   Mediana: {datos.median():.0f}°C")
        print(f"   Media: {datos.mean():.0f}°C")
        print(f"   N = {len(datos)} compuestos")

    # Grupo con mayor mediana
    grupo_max = plot_df.groupby(col_grupo)[col_PE].median().idxmax()
    mediana_max = plot_df.groupby(col_grupo)[col_PE].median().max()

    print("\n" + "="*50)
    print("CONCLUSIÓN:")
    print("="*50)
    print(f"✓ Mayor punto de ebullición: {grupo_max} ({mediana_max:.0f}°C)")

else:
    print("❌ No se encontraron las columnas necesarias")
    print("Columnas disponibles:", df.columns.tolist())
```

```python
# ============================================
# CORRELACIÓN PM vs PE POR GRUPO FUNCIONAL -
# ============================================

import plotly.express as px
import pandas as pd
import numpy as np

# Configuración (MODIFICA SEGÚN TUS COLUMNAS)
COL_PM = 'PMS'          # Peso Molecular
COL_PE = 'PES'          # Punto de Ebullición
COL_GRUPO = 'GF'        # Grupo Funcional

# Calcular R² por grupo
resultados = []
for grupo in df[COL_GRUPO].unique():
    subset = df[df[COL_GRUPO] == grupo].dropna(subset=[COL_PM, COL_PE])
    if len(subset) >= 3:
        r2 = subset[COL_PM].corr(subset[COL_PE])**2
        resultados.append({'Grupo': grupo, 'R²': r2, 'N': len(subset)})

df_r2 = pd.DataFrame(resultados).sort_values('R²', ascending=False)

# GRÁFICO 1: Barras de R² por grupo
fig1 = px.bar(df_r2, x='Grupo', y='R²', title='Correlación PM vs PE por Grupo Funcional',
              color='R²', color_continuous_scale='RdYlGn', range_color=[0,1],
              text='R²', labels={'R²': 'R² (correlación)'})
fig1.update_traces(texttemplate='%{text:.3f}', textposition='outside')
fig1.update_layout(width=900, height=500, xaxis_tickangle=-45)
fig1.show()

# GRÁFICO 2: Dispersión PM vs PE coloreado por grupo
fig2 = px.scatter(df, x=COL_PM, y=COL_PE, color=COL_GRUPO,
                  title='PM vs PE por Grupo Funcional',
                  labels={COL_PM: 'Peso Molecular (g/mol)', COL_PE: 'Punto de Ebullición (°C)'},
                  trendline='ols',
                  hover_data=[COL_GRUPO])
fig2.update_layout(width=900, height=550)
fig2.show()

# Mostrar resultados
print("\n📊 R² por grupo funcional:")
print(df_r2.to_string(index=False))
```

```python
# ============================================
# ANALISIS 3: COMPLETO Y CORREGIDO
# ============================================

import plotly.graph_objects as go
import numpy as np

# ============================================
# CONFIGURACIÓN (MODIFICABLE)
# ============================================
COL_GRUPO = 'GF'        # Columna de grupos funcionales
COL_PKA = 'pKa S'       # Columna de pKa
PKA_MIN = 0             # Mínimo pKa a incluir
PKA_MAX = 50            # Máximo pKa a incluir
MIN_COMPUESTOS = 2      # Mínimo compuestos por grupo

# ============================================
# PREPARAR DATOS
# ============================================
plot_df = df[[COL_GRUPO, COL_PKA]].dropna()
plot_df = plot_df[(plot_df[COL_PKA] >= PKA_MIN) & (plot_df[COL_PKA] <= PKA_MAX)]

# Filtrar grupos con mínimo de compuestos
grupos_con_datos = plot_df[COL_GRUPO].value_counts()
grupos_validos = grupos_con_datos[grupos_con_datos >= MIN_COMPUESTOS].index
plot_df = plot_df[plot_df[COL_GRUPO].isin(grupos_validos)]

print(f"✅ Compuestos válidos: {len(plot_df)}")
print(f"✅ Grupos funcionales: {plot_df[COL_GRUPO].nunique()}")

if len(plot_df) >= 4:
    # ============================================
    # CALCULAR ESTADÍSTICAS
    # ============================================
    medias = plot_df.groupby(COL_GRUPO)[COL_PKA].mean().sort_values(ascending=False)
    errores = plot_df.groupby(COL_GRUPO)[COL_PKA].std() / np.sqrt(plot_df.groupby(COL_GRUPO)[COL_PKA].count())

    # ============================================
    # CREAR GRÁFICO
    # ============================================
    fig = go.Figure()

    # Barras con barras de error
    fig.add_trace(go.Bar(
        x=medias.index.tolist(),
        y=medias.values,
        name='pKa promedio',
        marker_color='coral',
        marker_line_color='black',
        marker_line_width=1.5,
        error_y=dict(
            type='data',
            array=errores[medias.index].values,
            visible=True,
            color='darkred'
        ),
        text=medias.values.round(1),
        textposition='outside',
        textfont=dict(size=11, color='black')
    ))

    # Línea de referencia (amidas)
    fig.add_hline(
        y=16,
        line_dash="dash",
        line_color="red",
        line_width=2,
        annotation_text="Referencia (Amidas, pKa≈16)",
        annotation_position="top right"
    )

    # Diseño
    fig.update_layout(
        title=dict(
            text='<b>Análisis 3: Acidez y Estabilidad por Resonancia</b>',
            x=0.5
        ),
        xaxis=dict(
            title=dict(text=f'<b>{COL_GRUPO} - Grupo Funcional</b>'),
            tickangle=-45
        ),
        yaxis=dict(
            title=dict(text=f'<b>{COL_PKA} (pKa)</b>'),
            gridcolor='lightgray'
        ),
        width=900,
        height=600,
        plot_bgcolor='white'
    )

    fig.show()

    # ============================================
    # MOSTRAR RESULTADOS
    # ============================================
    print("\n" + "="*50)
    print("📊 pKa PROMEDIO POR GRUPO FUNCIONAL")
    print("="*50)
    for grupo, valor in medias.items():
        error_val = errores[grupo]
        print(f"  {grupo}: {valor:.1f} ± {error_val:.2f}")

    # Interpretación
    print("\n" + "="*50)
    print("🔬 INTERPRETACIÓN QUÍMICA")
    print("="*50)
    mas_acido = medias.idxmin()
    menos_acido = medias.idxmax()
    print(f"✅ MÁS ácido (pKa bajo): {mas_acido} ({medias[mas_acido]:.1f})")
    print(f"✅ MENOS ácido (pKa alto): {menos_acido} ({medias[menos_acido]:.1f})")

else:
    print(f"⚠️ Datos insuficientes: {len(plot_df)} compuestos (mínimo 4)")
```

```python
# ============================================
# ANALISIS : Criticidad de Solventes por Reacción - VERSIÓN FLEXIBLE
# ============================================

import plotly.express as px
import pandas as pd

# --------------------------------------------
# BLOQUE 1: ENCONTRAR COLUMNAS AUTOMÁTICAMENTE
# --------------------------------------------
col_reaccion = None
col_solvente = None

for col in df.columns:
    if 'Reaccion' in col or 'Reacción' in col or 'Tipo_Reaccion' in col or 'TR' in col or 'Nombre_Reaccion' in col:
        col_reaccion = col
    if 'Solvente' in col or 'S1' in col or 'Nombre_Solvente' in col:
        col_solvente = col

print(f"🔍 Columna Reacción encontrada: {col_reaccion}")
print(f"🔍 Columna Solvente encontrada: {col_solvente}")

# --------------------------------------------
# BLOQUE 2: PARÁMETROS MODIFICABLES
# --------------------------------------------
PESO_ALTA = 3      # Peso para solventes de alta toxicidad
PESO_MEDIA = 2     # Peso para solventes de media toxicidad
PESO_BAJA = 1      # Peso para solventes de baja toxicidad
MIN_REACCIONES = 2 # Mínimo de reacciones para considerar
CRITICO_ALTO = 2.0    # Umbral para criticidad muy alta
CRITICO_MEDIO = 1.5   # Umbral para criticidad alta

# Listas de solventes por toxicidad (MODIFICABLE según tu archivo)
SOLVENTES_ALTA = [
    'benceno', 'c6h6', 'tolueno', 'c7h8', 'xileno', 'c8h10',
    'cloroformo', 'chcl3', 'diclorometano', 'ch2cl2', 'cloruro de metileno',
    'clorobenceno', '1.2-dicloroetano', 'c2h4cl2'
]

SOLVENTES_MEDIA = [
    'eter', 'eter dietilico', 'c4h10o', 'thf', 'tetrahidrofurano', 'c4h8o', 'dioxano',
    'acetona', 'c3h6o', 'acetonitrilo', 'ch3cn', 'dmf', 'dimetilformamida', 'c3h7no',
    'hexano', 'c6h14', 'ciclohexano', 'c6h12'
]

SOLVENTES_BAJA = [
    'agua', 'h2o', 'etanol', 'c2h6o', 'etoh', 'metanol', 'ch3oh', 'meoh', 'isopropanol'
]

# --------------------------------------------
# BLOQUE 3: FUNCIÓN PARA CLASIFICAR TOXICIDAD
# --------------------------------------------
def clasificar_toxicidad(solvente):
    """Clasifica un solvente en niveles de toxicidad"""
    s = str(solvente).lower().strip()
    if s in SOLVENTES_BAJA:
        return 'Baja'
    elif s in SOLVENTES_MEDIA:
        return 'Media'
    elif s in SOLVENTES_ALTA:
        return 'Alta'
    else:
        return 'Desconocida'

# --------------------------------------------
# BLOQUE 4: VERIFICAR Y PREPARAR DATOS
# --------------------------------------------
if col_reaccion and col_solvente:
    plot_df = df[[col_reaccion, col_solvente]].dropna().copy()

    # Limpiar datos
    plot_df = plot_df[~plot_df[col_solvente].astype(str).str.contains('nan|None|0', na=False, case=False)]
    plot_df = plot_df[~(plot_df[col_solvente].astype(str).str.strip() == '')]
    plot_df[col_reaccion] = plot_df[col_reaccion].astype(str).str.strip()
    plot_df = plot_df[plot_df[col_reaccion] != '']

    # Clasificar toxicidad
    plot_df['Toxicidad'] = plot_df[col_solvente].apply(clasificar_toxicidad)

    print(f"✅ Reacciones analizadas: {len(plot_df)}")

    # --------------------------------------------
    # BLOQUE 5: CALCULAR PORCENTAJES Y PUNTAJE
    # --------------------------------------------
    # Porcentaje por reacción
    toxicidad_por_reaccion = plot_df.groupby(col_reaccion)['Toxicidad'].value_counts(normalize=True).unstack(fill_value=0) * 100

    # Asegurar que todas las categorías existan
    for nivel in ['Baja', 'Media', 'Alta', 'Desconocida']:
        if nivel not in toxicidad_por_reaccion.columns:
            toxicidad_por_reaccion[nivel] = 0

    # Calcular puntaje de criticidad
    toxicidad_por_reaccion['Puntaje_Criticidad'] = (
        toxicidad_por_reaccion['Alta'] * PESO_ALTA +
        toxicidad_por_reaccion['Media'] * PESO_MEDIA +
        toxicidad_por_reaccion['Baja'] * PESO_BAJA
    ) / 100

    # Ordenar por puntaje
    toxicidad_por_reaccion = toxicidad_por_reaccion.sort_values('Puntaje_Criticidad', ascending=False)

    # Filtrar reacciones con mínimo de ocurrencias
    conteo_reacciones = plot_df[col_reaccion].value_counts()
    reacciones_validas = conteo_reacciones[conteo_reacciones >= MIN_REACCIONES].index
    toxicidad_por_reaccion = toxicidad_por_reaccion[toxicidad_por_reaccion.index.isin(reacciones_validas)]

    print(f"✅ Reacciones válidas (mínimo {MIN_REACCIONES} ocurrencias): {len(toxicidad_por_reaccion)}")

    if len(toxicidad_por_reaccion) >= 2:
        # Tomar top 10
        top_criticas = toxicidad_por_reaccion.head(10)

        # --------------------------------------------
        # BLOQUE 6: GRÁFICO 1 - BARRAS APILADAS (MODIFICABLE)
        # --------------------------------------------
        fig1 = px.bar(
            top_criticas,
            x=top_criticas.index,
            y=['Alta', 'Media', 'Baja', 'Desconocida'],
            title='Reacciones más Críticas por Toxicidad de Solventes',
            labels={'value': 'Porcentaje de Uso (%)', 'variable': 'Nivel de Toxicidad', 'x': 'Tipo de Reacción'},
            color_discrete_map={
                'Alta': '#e74c3c',      # Rojo
                'Media': '#f39c12',     # Naranja
                'Baja': '#2ecc71',      # Verde
                'Desconocida': '#95a5a6' # Gris
            },
            barmode='relative'
        )

        fig1.update_layout(
            width=900,
            height=550,
            title_x=0.5,
            xaxis_tickangle=-45,
            yaxis_title='Porcentaje (%)',
            yaxis_range=[0, 100],
            plot_bgcolor='white',
            showlegend=True,
            legend_title='Nivel de Toxicidad'
        )

        fig1.show()

        # --------------------------------------------
        # BLOQUE 7: GRÁFICO 2 - PUNTAJE DE CRITICIDAD (MODIFICABLE)
        # --------------------------------------------
        # Crear colores según criticidad
        colores_criticos = [
            '#e74c3c' if p > CRITICO_ALTO
            else '#f39c12' if p > CRITICO_MEDIO
            else '#2ecc71'
            for p in top_criticas['Puntaje_Criticidad']
        ]

        fig2 = px.bar(
            top_criticas,
            y=top_criticas.index,
            x='Puntaje_Criticidad',
            orientation='h',
            title='Puntaje de Criticidad por Reacción',
            labels={'Puntaje_Criticidad': 'Puntaje de Criticidad (0-3)', 'y': 'Tipo de Reacción'},
            text='Puntaje_Criticidad'
        )

        fig2.update_traces(
            marker_color=colores_criticos,
            marker_line_color='black',
            marker_line_width=0.5,
            texttemplate='%{text:.2f}',
            textposition='outside'
        )

        fig2.update_layout(
            width=800,
            height=550,
            title_x=0.5,
            plot_bgcolor='white',
            showlegend=False,
            xaxis_range=[0, 3]
        )

        # Añadir líneas verticales de referencia
        fig2.add_vline(
            x=CRITICO_MEDIO,
            line_dash="dash",
            line_color="#f39c12",
            line_width=2,
            annotation_text=f"Crítico ({CRITICO_MEDIO})",
            annotation_position="top"
        )
        fig2.add_vline(
            x=CRITICO_ALTO,
            line_dash="dash",
            line_color="#e74c3c",
            line_width=2,
            annotation_text=f"Muy Crítico ({CRITICO_ALTO})",
            annotation_position="top"
        )

        fig2.show()

        # --------------------------------------------
        # BLOQUE 8: MOSTRAR RESULTADOS EN CONSOLA
        # --------------------------------------------
        print("\n" + "="*50)
        print("RANKING DE REACCIONES POR CRITICIDAD")
        print("="*50)
        for i, (reaccion, row) in enumerate(top_criticas.iterrows(), 1):
            print(f"  {i}. {reaccion}: Puntaje={row['Puntaje_Criticidad']:.2f}")

        # Alertas para reacciones peligrosas
        peligrosas = top_criticas[top_criticas['Alta'] > 50]
        if len(peligrosas) > 0:
            print("\n" + "="*30)
            print("⚠️ ALERTA: Reacciones con uso de solventes de ALTA toxicidad >50%")
            print("="*30)
            for reaccion in peligrosas.index:
                print(f"  {reaccion}: {peligrosas.loc[reaccion, 'Alta']:.1f}%")

        # Resumen general
        print("\n" + "="*50)
        print("RESUMEN GENERAL")
        print("="*50)
        print(f"  Total de reacciones analizadas: {len(plot_df[col_reaccion].unique())}")
        print(f"  Reacciones críticas (puntaje > {CRITICO_ALTO}): {len(top_criticas[top_criticas['Puntaje_Criticidad'] > CRITICO_ALTO])}")
        print(f"  Reacciones de alta criticidad (puntaje > {CRITICO_MEDIO}): {len(top_criticas[top_criticas['Puntaje_Criticidad'] > CRITICO_MEDIO])}")

    else:
        print(f"❌ Datos insuficientes: {len(toxicidad_por_reaccion)} reacciones (mínimo 2)")

else:
    print("❌ No se encontraron las columnas necesarias")
    print("Columnas disponibles:", df.columns.tolist())
```

```python
# ============================================
# ANALISIS: PM Producto por Tipo de Reacción - VERSIÓN FLEXIBLE
# ============================================

import plotly.express as px
import pandas as pd

# --------------------------------------------
# BLOQUE 1: ENCONTRAR COLUMNAS AUTOMÁTICAMENTE
# --------------------------------------------
col_tipo_reaccion = None
col_pm_producto = None

for col in df.columns:
    if 'Tipo_Reaccion' in col or 'Tipo' in col or 'TR' in col:
        col_tipo_reaccion = col
    if 'PM_Producto' in col or 'PMP' in col or 'PM_Producto' in col:
        col_pm_producto = col

print(f"🔍 Columna Tipo de Reacción encontrada: {col_tipo_reaccion}")
print(f"🔍 Columna PM Producto encontrada: {col_pm_producto}")

# --------------------------------------------
# BLOQUE 2: PARÁMETROS MODIFICABLES
# --------------------------------------------
PM_MIN = 0          # Mínimo PM del producto
PM_MAX = 600        # Máximo PM del producto
MIN_REACCIONES = 3  # Mínimo de reacciones por tipo
BAR_COLOR = 'seagreen'  # Color de las barras

if col_tipo_reaccion and col_pm_producto:
    # --------------------------------------------
    # BLOQUE 3: PREPARAR Y FILTRAR DATOS
    # --------------------------------------------
    plot_df = df[[col_tipo_reaccion, col_pm_producto]].dropna().copy()

    # Filtrar por rango de PM
    plot_df = plot_df[(plot_df[col_pm_producto] >= PM_MIN) & (plot_df[col_pm_producto] <= PM_MAX)]

    # Limpiar tipos de reacción
    plot_df = plot_df[~plot_df[col_tipo_reaccion].astype(str).isin(['nan', 'None', ''])]
    plot_df[col_tipo_reaccion] = plot_df[col_tipo_reaccion].astype(str).str.strip()
    plot_df = plot_df[plot_df[col_tipo_reaccion] != '']

    # Filtrar tipos con mínimo de reacciones
    tipos_validos = plot_df[col_tipo_reaccion].value_counts()
    tipos_validos = tipos_validos[tipos_validos >= MIN_REACCIONES].index
    plot_df = plot_df[plot_df[col_tipo_reaccion].isin(tipos_validos)]

    print(f"✅ Tipos de reacción analizados: {list(plot_df[col_tipo_reaccion].unique())}")
    print(f"✅ Total reacciones: {len(plot_df)}")

    if len(plot_df) >= 5:
        # --------------------------------------------
        # BLOQUE 4: CALCULAR PROMEDIOS
        # --------------------------------------------
        medias = plot_df.groupby(col_tipo_reaccion)[col_pm_producto].mean().sort_values(ascending=False)

        print(f"\n📊 PM Promedio por tipo de reacción:")
        for tipo, pm in medias.items():
            print(f"   {tipo}: {pm:.0f} g/mol")

        # --------------------------------------------
        # BLOQUE 5: CREAR GRÁFICO DE BARRAS (MODIFICABLE)
        # --------------------------------------------
        fig = px.bar(
            x=medias.index,
            y=medias.values,
            title='PM Promedio del Producto por Tipo de Reacción',
            labels={'x': 'Tipo de Reacción', 'y': 'PM Promedio del Producto (g/mol)'},
            text=medias.values,
            color=medias.values,
            color_continuous_scale=['lightgreen', 'darkgreen']
        )

        fig.update_traces(
            texttemplate='%{text:.0f}',
            textposition='outside',
            marker_color=BAR_COLOR,
            marker_line_color='black',
            marker_line_width=1,
            opacity=0.8
        )

        fig.update_layout(
            width=900,
            height=550,
            title_x=0.5,
            xaxis_tickangle=-45,
            yaxis_title='PM Promedio (g/mol)',
            plot_bgcolor='white',
            showlegend=False
        )

        fig.add_hline(
            y=medias.mean(),
            line_dash="dash",
            line_color="gray",
            annotation_text=f"Promedio global: {medias.mean():.0f} g/mol",
            annotation_position="top right"
        )

        fig.show()

        # --------------------------------------------
        # BLOQUE 6: CONCLUSIONES
        # --------------------------------------------
        print("\n" + "="*50)
        print("CONCLUSIONES DEL ANÁLISIS 6")
        print("="*50)

        tipo_max = medias.index[0]
        tipo_min = medias.index[-1]

        print(f"✓ Tipo con mayor PM promedio: {tipo_max} ({medias[tipo_max]:.0f} g/mol)")
        print(f"✓ Tipo con menor PM promedio: {tipo_min} ({medias[tipo_min]:.0f} g/mol)")
        print(f"✓ Diferencia entre máximo y mínimo: {medias[tipo_max] - medias[tipo_min]:.0f} g/mol")

        # Interpretación según tipos comunes
        if 'Condensación' in medias.index or 'Cond' in medias.index:
            print("✓ Las reacciones de condensación generan productos más pesados")
        if 'Reducción' in medias.index or 'Red' in medias.index:
            print("✓ Las reacciones de reducción generan productos más ligeros")
        if 'Sustitución' in medias.index or 'SN' in medias.index:
            print("✓ Las reacciones de sustitución tienen PM intermedios")

    else:
        print(f"❌ Datos insuficientes: {len(plot_df)} reacciones (mínimo 5)")
else:
    print("❌ No se encontraron las columnas necesarias")
    print("Columnas disponibles:", df.columns.tolist())
```

```python
import plotly.graph_objects as go
import pandas as pd

def clasificar(s):
    if pd.isna(s): return None
    s = str(s).lower()
    if 'agua' in s: return 'Agua (Verde)'
    if 'etanol' in s or 'metanol' in s: return 'Alcohol (Verde)'
    if 'thf' in s: return 'Eter (Moderado)'
    if 'dmf' in s or 'diclorometano' in s: return 'Aprotico Polar (Moderado)'
    if 'benceno' in s or 'tolueno' in s: return 'Aromatico/Halogenado (Toxico)'
    return 'Otros'

# Recolectar datos de C1 y S1
datos = []
for _, row in df.iterrows():
    medio = row.get('Med')
    if pd.notna(medio):
        for col in ['C1', 'S1']:
            if col in df.columns and pd.notna(row[col]):
                tipo = clasificar(row[col])
                if tipo:
                    datos.append({'Med': medio, 'Tipo': tipo})

plot_df = pd.DataFrame(datos)
cross = pd.crosstab(plot_df['Med'], plot_df['Tipo'], normalize='index') * 100

fig = go.Figure()
for col in cross.columns:
    fig.add_trace(go.Bar(name=col, x=cross.index, y=cross[col],
                         text=cross[col].round(0), textposition='inside'))

fig.update_layout(barmode='stack', title='Solventes por Tipo de Reacción',
                  xaxis_title='Medio', yaxis_title='% Uso', width=800, height=500)
fig.show()
print(cross.round(1))
```
